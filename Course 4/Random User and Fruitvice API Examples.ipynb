```python
# Defining various properties of a user using pandas

def get_users():
    users =[]
     
    for user in RandomUser.generate_users(10):
        users.append({"Name":user.get_full_name(),"Gender":user.get_gender(),"City":user.get_city(),"State":user.get_state(),"Email":user.get_email(), "DOB":user.get_dob(),"Picture":user.get_picture()})
      
    return pd.DataFrame(users)     

get_users()
```

```python
# A dataframe thats takes in the properties of the users

df1 = pd.DataFrame(get_users())  
```

# FRUITIVE API

```python
import requests
import json
```

```python
# We will obtain the fruitvice API data using requests.get("url") function. The data is in a json format

data = requests.get("https://fruityvice.com/api/fruit/all")
```

```python
# We will retrieve results using json.loads() function

results = json.loads(data.text)
```

```python
# We will convert our json data into pandas data frame

pd.DataFrame(results)
```

```python
# The result is in a nested json format. The 'nutrition' column contains multiple subcolumns,
# so the data needs to be 'flattened' or normalized

df2 = pd.json_normalize(results)
df2
```

```python
cherry = df2.loc[df2["name"] == 'Cherry']
(cherry.iloc[0]['family']) , (cherry.iloc[0]['genus'])
```

```python
# How many calories are contained in banana

banana_cal = df2.loc[df2["name"] == 'Banana']
banana_cal.iloc[0]['nutritions.calories']
```
